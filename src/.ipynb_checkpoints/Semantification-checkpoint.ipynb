```python
import spotlight
```

```python
text = 'President Obama on Monday will call for a new minimum tax rate for individuals'
```

```python
url = 'http://spotlight.sztaki.hu:2222/rest/annotate'
```

```python
import numpy as np
```

```python
def get_event_entry(data):
    """
    generator to iterate over events
    """
    for key, value in data.item().items():
        yield (key, value)
```

```python
data = np.load('../data/processed_data.npy')
```

```python
def get_spotlight_responses(event_data):
    data = []
    for i, (key, value) in enumerate(get_event_entry(event_data)):
        
        text = ' '.join(value)
        try:
            res = spotlight.annotate(url, text, confidence=0.7, support=50)
            
            if res:
                for entry in res:
                    data.append((key, res))
        except:
            pass

    return data
```

```python
get_spotlight_responses(data)
```

```python
np.save('../data/spotlight_responses', get_spotlight_responses(data))
```
